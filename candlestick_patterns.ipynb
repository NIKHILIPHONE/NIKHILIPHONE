```python
import pandas as pd

file_path = '/content/StockDataBANKBARODA_1.csv'

# Skip first 2 unnecessary rows
stock_data = pd.read_csv(file_path, skiprows=2)

# Manually rename columns
stock_data.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']

# Convert Date column to datetime safely
stock_data['Date'] = pd.to_datetime(stock_data['Date'], errors='coerce')

print("✅ Cleaned Data Preview:")
display(stock_data.head())
```

```python
if 'Date' not in stock_data.columns:
    raise ValueError("'Date' column is missing from the data.")

# Check for missing values
missing_values = stock_data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Handle missing values
stock_data.fillna(method='ffill', inplace=True)
stock_data.dropna(inplace=True)

# Ensure the columns for candlestick plot exist
required_columns = ['Open', 'High', 'Low', 'Close']
for column in required_columns:
    if column not in stock_data.columns:
        raise ValueError(f"'{column}' column is missing from the data.")

# Ensure the data is sorted by date
stock_data.sort_values('Date', inplace=True)
```
